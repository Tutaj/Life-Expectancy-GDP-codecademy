```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
print(df.head(20))
```

```python
#Has life expectancy increased over time in the six nations?
plt.figure(figsize = (12,7))
sns.lineplot(data = df, x = 'Year', y = 'Life expectancy at birth (years)', hue = 'Country', palette = ['#d53e4f','#fc8d59','#fee08b','#e6f598','#99d594','#3288bd'] )
plt.title('Life expectancy each year in 6 countries')
```

```python
#Has GDP increased over time in the six nations?
```

```python
#plt.figure(figsize = (12,7))
#sns.lineplot(data = df, x = 'Year', y = 'GDP', hue = 'Country', palette = ['#d53e4f','#fc8d59','#fee08b','#e6f598','#99d594','#3288bd'])
### this doesn't work because GDP has too large spread in other countries to see difference in Zimbabwe, Chile etc.

def gdp_increase(country):
    plt.figure(figsize = (12,7))
    sns.lineplot(data = df[df.Country == country], x = 'Year', y = 'GDP')
    plt.title(country + "'s GDP over time")
gdp_increase('Mexico')
```

```python
#Is there a correlation between GDP and life expectancy of a country?
```

```python
def gdp_le_correlation(country):
    plt.figure(figsize = (12,7))
    sns.regplot(data = df[df.Country == country], x = 'Life expectancy at birth (years)', y = 'GDP', line_kws = {'color': 'yellow'})
    plt.title(country + "'s GDP vs Life Expectancy Correlation")
gdp_le_correlation('Germany')
#looks like there is strong correlation between GDP and Life expectancy
```

```python
#What is the average life expectancy in these nations?
plt.figure(figsize = (12,7))
sns.barplot()
plt.title('Average Life Expectancy')
```

```python
#What is the distribution of that life expectancy?
#this is with zimbabwe which makes data hard to read
colors = ['#8c510a','#d8b365','#f6e8c3','#c7eae5','#5ab4ac','#01665e']
plt.figure(figsize = (12,7))
sns.violinplot(data = df, x = 'Country', y = 'Life expectancy at birth (years)', palette = colors)
plt.title('Distribution of Life Expectancy')
without_zimbabwe = df[df.Country != 'Zimbabwe']
plt.figure(figsize = (12,7))
sns.violinplot(data = without_zimbabwe, x = 'Country', y = 'Life expectancy at birth (years)', palette = colors)
plt.title('Distribution of Life Expectancy(Without Zimbabwe)')
```
